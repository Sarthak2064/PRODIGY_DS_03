```python
# Visualizing Categorical columns using Barplot

import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns you want to visualize
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome']

# Calculate the number of rows and columns for subplots
num_rows = len(cat_cols) // 2
num_cols = 2 if len(cat_cols) % 2 == 0 else 3  # Adjust the number of columns as needed

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten axes for easy iteration
axes = axes.flatten()

# Loop through categorical columns and create countplots
for i, feature in enumerate(cat_cols):
    sns.countplot(x=feature, data=bank_df, palette='Wistia', ax=axes[i])
    axes[i].set_title(f'Bar Plot of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=90)

# Remove any empty subplots if the number of columns is odd
if len(cat_cols) % 2 != 0:
    fig.delaxes(axes[-1])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Plotting BoxPlot and Checking for Outliers
bank_df.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='#7b3f00')
plt.show()
```

```python
# Removing outliers

column = bank_df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
bank_df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
# Plotting boxplot after removing outliers
bank_df.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='#808000')
plt.show()
```

```python
# Checking for correlation using Correlation Plot
corr = bank_df.corr()
print(corr)
corr = corr[abs(corr)>=0.90]
sns.heatmap(corr,annot=True,cmap='Set3',linewidths=0.2)
plt.show()
```

```python
# Feature Selection using Correlation
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
# copy with original dataframe
bank = bank_df.copy()
bank.columns
```

```python
# Removing high correlated columns from the dataset
bank.drop(high_corr_cols,inplace=True,axis=1)  # axis=1 indicates columns
bank.columns
```

```python
# showing dimensions of the updated dataset
bank.shape
```

```python
# Conversion of categorical columns into numerical columns using label encoder.
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
bank_df_encoded = bank.apply(lb.fit_transform)
bank_df_encoded
```

```python
bank_df_encoded['deposit'].value_counts()
```

```python
# Selecting Independent and Dependent Variables
x = bank_df_encoded.drop('deposit',axis=1)  # independent variable
y = bank_df_encoded['deposit']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
# Splitting the dataset into Train and Test datasets
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  # Training Accuracy
    print('Testing Score',test_score)    # Testing Accuracy
```

```python
# Importing  Decision Tree library
from sklearn.tree import DecisionTreeClassifier
```

```python
# Building Decision Tree Classifier Model
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
# Evaluating training and testing accuracy
mscore(dt)
```

```python
# Generating prediction
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
# # Evaluate the model - confusion matrix, classification Report, Accuaracy
eval_model(y_test,ypred_dt)
```
